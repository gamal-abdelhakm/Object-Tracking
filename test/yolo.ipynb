```python
from ultralytics import YOLO
import cv2

model = YOLO('yolo11n.pt')
```

```python
img = cv2.imread(r'E:\Photos\4.jpg')
results = model(img)
cv2.imshow('result',results[0].plot())
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
model = YOLO('yolo11n.pt')
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    results = model.track(frame, classes=0)
    cv2.imshow('result',results[0].plot())
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
